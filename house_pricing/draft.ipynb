```python
import pandas as pd
```

```python
# save filepath to variable for easier access
melbourne_file_path = "data/raw/melb_data.csv"
# read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path)
# print a summary of the data in Melbourne data
melbourne_data.describe()
```

```python
melbourne_data = melbourne_data.dropna(axis=0)  # rows with missing values are removed
```

```python
y = melbourne_data.Price
```

```python
melbourne_features = [
    "Rooms",
    "Bathroom",
    "Landsize",
    "BuildingArea",
    "YearBuilt",
    "Lattitude",
    "Longtitude",
]
```

```python
X = melbourne_data[melbourne_features]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
melbourne_model = DecisionTreeRegressor(random_state=1)
```

```python
melbourne_model.fit(X=X, y=y)
```

```python
type(melbourne_model), type(linear_model), linear_model.get_params()
```

```python
melbourne_model.get_params()
```

```python
predicted_prices = melbourne_model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y, predicted_prices)
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

melbourne_model = DecisionTreeRegressor()

melbourne_model.fit(train_X, train_y)

val_predictions = melbourne_model.predict(val_X)

mean_absolute_error(val_y, val_predictions)
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return mae
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(
        "Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" % (max_leaf_nodes, my_mae)
    )
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melbs_preds = forest_model.predict(val_X)
mean_absolute_error(val_y, melbs_preds)
```
